## Let's perform a simple EDA (Exploratory Data Analysis) on the wine quality dataset using pandas for basic statistics and seaborn for visualization to understand the data's distribution and relationships.

### First things first, imports.

```python
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.lines import Line2D
import pandas as pd
import os
import sys
```

## Data Loading 
### We load the data into a data frame. DataFrames simplify data manipulation and analysis.

```python
# Load the dataset again with the correct separator
data_df = pd.read_csv('/Users/yassine/Documents/BachelorArbeit/aperion/wine_prediction_project/data/winequality-white.csv', sep=';')
```

## Data Cleaning
### We clean the data using the `datacleaning` module. This represents the first data pre-processing step and involves a first data analysis to check if there are inconsistencies with the data, e.g. correcting value anomalies like 999 instead of 0.999

```python
sys.path.insert(0, '../src')
import datacleaning
data = datacleaning.clean_data(data_df)
```

## Visualizing the Data Distribution
### This will help us understand the distributions of samples we have, which is crucial to understand and analyse the data later, since the data might not provide a comprehensive representation, for quality scores with small sample size.

```python
sys.path.insert(0, '../src')
import eda

eda.plot_distribution(data,"quality")
```

## Basic Statistical Overview
### We utilize the data.describe() method from pandas to obtain a basic statistical summary of our dataset. This summary includes measures such as mean, standard deviation, minimum, and maximum values, along with the 25th, 50th (median), and 75th percentiles for each numerical column in the DataFrame. The resulting output, stored in basic_stats, offers a quick overview of the data's distribution and central tendencies.

```python
# Display basic statistics
basic_stats = data.describe()
basic_stats
```

## Analyzing Feature Distribution Across Wine Quality Ratings
### To uncover insights into how specific attributes correlate with wine quality, we leverage the `plot_feature_by_quality` function, enabling a focused analysis of feature distribution within various quality brackets.

```python
sys.path.insert(0, '../src')
import eda

# We first extract all features form the data
features = [column for column in data.columns if column != "quality"]

# And then plot a graph for each feature
for feature in features:   
    eda.plot_feature_by_quality(data=data, feature=feature)

# The loop can be replaced by single feature arguments.
```

## Visualizing Correlations with a Heatmap
### Utilizing the `plot_correlation_matrix` function, we construct a heatmap to visualize variable interdependencies. Values closer to **1** signify a strong positive correlation, values closer to **-1** indicate a strong negative correlation, and values around **0** suggest little to no correlation. This facilitates the identification of significant relationships crucial for further analysis.

### The data from the heat map consolidates the information obtained from the previous graphs about the underlying patterns and correlations

```python
sys.path.insert(0, '../src')
import eda

eda.plot_correlation_matrix(data)
```
